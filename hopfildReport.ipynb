```python
shapes = []
for fileName in learnShapesFiles:
    shape = imp.parse_image_to_shape(fileName)
    if len(shape) == SIZE ** 2:
        print("FIND SHAPE = \"" + fileName + "\"")
        shapes.append(shape)
    else:
        print("FILE IS NOT SHAPE = \"" + fileName + "\"")
```

```python
builder = nlb.NeyronLayerBuilder(SIZE ** 2)
for shape in shapes:
    builder.teach(shape)
        
layer = builder.build()
```
